```python
import pandas as pd
```

```python
# Separate 7 models and human data. 
# Assume Human text is from model 100,000 and machine text is from model 7 in domain 1
file_path_1 = 'data/domain1_train.json'
df1 = pd.read_json(file_path_1, lines=True)
df1.loc[df1['label'] == 1, 'model'] = 100000
df1.loc[df1['label'] == 0, 'model'] = 7

file_path_2 = 'data/domain2_train.json'
df2 = pd.read_json(file_path_2, lines=True)
df2['model'].fillna(100000, inplace=True)

df_comb = pd.concat([df1, df2],axis=0,ignore_index=True)

display(df_comb.groupby(df_comb['model']).count())
```

```python
# Separate observations per model
m0 = df_comb.loc[df_comb['model']==0]
m1 = df_comb.loc[df_comb['model']==1]
m2 = df_comb.loc[df_comb['model']==2]
m3 = df_comb.loc[df_comb['model']==3]
m4 = df_comb.loc[df_comb['model']==4]
m5 = df_comb.loc[df_comb['model']==5]
m6 = df_comb.loc[df_comb['model']==6]
m7 = df_comb.loc[df_comb['model']==7]
mh = df_comb.loc[df_comb['model']==100000]
```

# Create down-sampled weighted dataset

```python
# majority_size = int(len(mh) / 4)
# # Randomly oversample the minority class to match the majority class size
# oversampled_minority_m0 = m0.sample(n=majority_size, replace=True)
# oversampled_minority_m1 = m1.sample(n=majority_size, replace=True)
# oversampled_minority_m2 = m2.sample(n=majority_size, replace=True)
# oversampled_minority_m3 = m3.sample(n=majority_size, replace=True)

# df_ds_weighted = pd.concat([mh, oversampled_minority_m0, oversampled_minority_m1, oversampled_minority_m2, oversampled_minority_m3])

df_ds_weighted = pd.concat([mh, m0, m1, m2, m3])

df_ds_weighted =  df_ds_weighted.drop('model', axis=1)

display(df_ds_weighted.head())
display(df_ds_weighted.tail())
print(f"This dataset consist 11900 human text and {df_ds_weighted.shape[0] - 11900} machine text")

```

```python
df_ds_weighted.to_csv('cut_data.csv', index=False)
```

# Create Balanced Dataset

```python
majority_size = len(mh)
# Randomly oversample the minority class to match the majority class size
oversampled_minority_m0 = m0.sample(n=majority_size, replace=True)
oversampled_minority_m1 = m1.sample(n=majority_size, replace=True)
oversampled_minority_m2 = m2.sample(n=majority_size, replace=True)
oversampled_minority_m3 = m3.sample(n=majority_size, replace=True)
oversampled_minority_m4 = m4.sample(n=majority_size, replace=True)
oversampled_minority_m5 = m5.sample(n=majority_size, replace=True)
oversampled_minority_m6 = m6.sample(n=majority_size, replace=True)
oversampled_minority_m7 = m7.sample(n=majority_size, replace=True)

# Combine the oversampled minority class with the majority class
df_balanced = pd.concat([mh, oversampled_minority_m0, oversampled_minority_m1, oversampled_minority_m2,
                         oversampled_minority_m3, oversampled_minority_m4, oversampled_minority_m5,
                         oversampled_minority_m6, oversampled_minority_m7])

df_balanced =  df_balanced.drop('model', axis=1)

display(df_balanced.head())
display(df_balanced.tail())
df_balanced.shape
```

```python
df_balanced.to_csv('balanced_data.csv', index=False)
```

# Create Up-sampled weighted dataset

```python
# Randomly oversample the minority class to match the majority class size
oversampled_minority_m0 = m0.sample(n=majority_size*2, replace=True)
oversampled_minority_m1 = m1.sample(n=majority_size*2, replace=True)
oversampled_minority_m2 = m2.sample(n=majority_size*2, replace=True)
oversampled_minority_m3 = m3.sample(n=majority_size*2, replace=True)
oversampled_minority_m4 = m4.sample(n=majority_size, replace=True)
oversampled_minority_m5 = m5.sample(n=majority_size, replace=True)
oversampled_minority_m6 = m6.sample(n=majority_size, replace=True)
oversampled_minority_m7 = m7.sample(n=majority_size, replace=True)

# Combine the oversampled minority class with the majority class
df_weighted = pd.concat([mh, oversampled_minority_m0, oversampled_minority_m1, oversampled_minority_m2,
                         oversampled_minority_m3, oversampled_minority_m4, oversampled_minority_m5,
                         oversampled_minority_m6, oversampled_minority_m7])

df_weighted = df_weighted.drop('model', axis=1)

display(df_weighted.head())
display(df_weighted.tail())
print(f"model 0,1,2,3 have weight of {len(oversampled_minority_m0)/len(df_weighted)}")
print(f"model 4,5,6,7 have weight of {len(oversampled_minority_m4)/len(df_weighted)}")
df_weighted.shape
```

```python
df_weighted.to_csv('weighted_data.csv', index=False)
```
